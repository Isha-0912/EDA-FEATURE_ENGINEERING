```python
df.dropna(subset=["Type","Android Ver","Current Ver","Content Rating"],inplace=True)
df.isnull().sum()
```

```python
mean=df["Rating"].mean()
df.fillna({"Rating": mean},inplace=True)
df.isnull().sum()
```

```python
df["Reviews"].unique()
df["Reviews"]=pd.to_numeric(df["Reviews"])
```

```python
df["Size"].unique()
df["Price"].unique()
```

```python
df["Price"]=df["Price"].str.replace(r'[\$]','',regex=True)
df["Price"]=pd.to_numeric(df["Price"])
```

```python
df["Installs"].unique()
df["Installs"]=df["Installs"].str.replace(r'[\+,]','',regex=True)
df["Installs"]=pd.to_numeric(df["Installs"])
df.duplicated().sum()
```

```python
df["Size"].value_counts()
df["Size"]=df["Size"].replace("Varies with device",np.nan)
```

```python
def convert_size(size):
    if pd.isna(size):
        return np.nan

    if "M" in size:
        return float(size.replace("M", ""))

    if "k" in size:
        return float(size.replace("k", "")) / 1024

    return np.nan
```

```python
df["Size"] = df["Size"].apply(convert_size)
df.info()
```

```python
df["Size"]=df["Size"].fillna(df["Size"].median())
df.isnull().sum()
```

```python
df["Last Updated"]=pd.to_datetime(df["Last Updated"])
df["Last_updated_year"]=df["Last Updated"].dt.year
df["Last_updated_month"]=df["Last Updated"].dt.month
```

```python
df.drop('Last Updated',axis=1,inplace=True)
df.info()
```

```python
df.drop(['App','Android Ver','Current Ver'],axis=1,inplace=True)
df.info()
```

```python
for col in df.columns:
    print(col)
    print("unique:",df[col].unique())
    print(df[col].value_counts().head())
```

```python
category = pd.get_dummies(
    df["Category"],
    drop_first=True,
    dtype=int
)
df = pd.concat([df, category], axis=1)
```

```python
df.drop("Category", axis=1, inplace=True)
```

```python
df["Genres"].nunique()
```

```python
genres = pd.get_dummies(
    df["Genres"],
    drop_first=True,
    dtype=int
)

df = pd.concat([df, genres], axis=1)

df.drop("Genres", axis=1, inplace=True)
```

```python
content_rating = pd.get_dummies(
    df["Content Rating"],
    drop_first=True,
    dtype=int
)
df = pd.concat(
    [df, content_rating],
    axis=1
)
df.drop(
    "Content Rating",
    axis=1,
    inplace=True
)
```
