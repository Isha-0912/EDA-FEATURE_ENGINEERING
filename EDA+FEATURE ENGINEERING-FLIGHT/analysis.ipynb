```python
df["Journey_day"]=pd.to_datetime(
    df["Date_of_Journey"],
    dayfirst=True
).dt.day
df["Journey_month"]=pd.to_datetime(
    df["Date_of_Journey"],
    dayfirst=True
).dt.month
df.drop("Date_of_Journey",axis=1,inplace=True)
df.head()
```

```python
df.columns
```

```python
df.head()
```

```python
df["Dep_Time"]=pd.to_datetime(df["Dep_Time"])
df["Dep_Hour"]=df["Dep_Time"].dt.hour
df["Dep_Min"]=df["Dep_Time"].dt.minute
```

```python
df.drop("Dep_Time",axis=1,inplace=True)
```

```python
df.columns
df.head()
```

```python
df["Arrival_Time"]=df["Arrival_Time"].str.split(" ").str[0]
df["Arrival_Time"]=pd.to_datetime(df["Arrival_Time"])
df["Arr_hour"]=df["Arrival_Time"].dt.hour
df["Arr_min"]=df["Arrival_Time"].dt.minute
```

```python
df.drop("Arrival_Time", axis=1, inplace=True)
```

```python
df.head(3)
```

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"]=df["Total_Stops"].map(
    {"non-stop":0,
     "1 stop":1,
     "2 stops":2,
     "3 stops":3,
     "4 stops":4
     }
)
```

```python
df.head(3)
```

```python
df["Total_Stops"].isnull().sum()
```

```python
df.select_dtypes(include="object").columns
```

```python
# Add missing hour values
df["Duration"] = df["Duration"].apply(
    lambda x: "0h " + x if "h" not in x else x
)

# Add missing minute values
df["Duration"] = df["Duration"].apply(
    lambda x: x + " 0m" if "m" not in x else x
)

# Extract hours
df["Duration_Hour"] = df["Duration"].str.extract(r"(\d+)h").astype(int)

# Extract minutes
df["Duration_Min"] = df["Duration"].str.extract(r"(\d+)m").astype(int)

# Drop original column
df.drop("Duration", axis=1, inplace=True)
```

```python
df.drop("Additional_Info", axis=1, inplace=True)
```

```python
df.drop("Route", axis=1, inplace=True)
```

```python
airline = pd.get_dummies(
    df["Airline"],
    drop_first=True,
    dtype=int
)
source = pd.get_dummies(
    df["Source"],
    drop_first=True,
    dtype=int
)
destination = pd.get_dummies(
    df["Destination"],
    drop_first=True,
    dtype=int
)
```

```python
df = pd.concat(
    [df, airline, source, destination],
    axis=1
)
```

```python
df.drop(
    ["Airline", "Source", "Destination"],
    axis=1,
    inplace=True
)
```
